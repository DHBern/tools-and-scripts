```python
# Mit diesem Code sollen alle einzelnen Berufe aus dem Jahr 1925 herausgefunden werden.

# Mit diesem Code werden die notwendigen Programmbibliotheken importiert.

import pandas as pd
import matplotlib.pyplot as plt

# Hier werden die vorhandenen TSV (Tab Seperated Value - also durch Leerschläge werden die einzelnen Werte getrennt) eingelesen. 
# Es handelt sich dabei um die Daten b3 und b6. Diese beinhalten alle Beiträge des Jahres 1925.

jobs_b3 =  pd.read_table ('C:/Users/dkilc/OneDrive - Universitaet Bern/Neu/Desktop/Studium/DH_WORK/Telefonbuch/b3.tsv')
jobs_b6 =  pd.read_table ('C:/Users/dkilc/OneDrive - Universitaet Bern/Neu/Desktop/Studium/DH_WORK/Telefonbuch/b6.tsv')
jobs_b3_1925 = jobs_b3[jobs_b3.Year == 1925]
jobs_b6_1925 = jobs_b6[jobs_b6.Year == 1925]

# Nun werden all diese Einträge in einem Datensatz zusammengefasst.

jobs_1925 = pd.concat([jobs_b3_1925, jobs_b6_1925], ignore_index=True)
```

```python
# Dieser Code zählt die Länge des Datensatzes, also wie viele Einträge sich darin befinden.

Entries_1925 = len(jobs_1925)
print("Für das Jahr 1925 haben sich insgesamt", Entries_1925, "eintragen lassen.")
```

```python
# Dieser Code zählt im Datensatze 1925 jene Zellen, die in der Spalte "Supplement" keinen Eintrag besitzen und rechnet den Prozentsatz aus.

empty_Entries_1925 = jobs_1925["Supplement"].isnull().sum()
percent_null = (empty_Entries_1925 / Entries_1925) * 100

print("Davon haben", empty_Entries_1925, "oder", percent_null, "% die Spalte 'Supplement' freigelassen.")
```

```python
# Dieser Code zählt im Datensatze 1925 jene Zellen, die in der Spalte "Supplement" einen Eintrag besitzen und rechnet den Prozentsatz aus.

notempty_Entries_1925 = jobs_1925["Supplement"].notnull().sum()
percent_non_null = (notempty_Entries_1925 / Entries_1925) * 100

print(notempty_Entries_1925, "oder rund", percent_non_null,  "% haben etwas in die Spalte 'Supplement eingetragen.")

# Dieser Code (value_counts_1925) zählt die individuellen Einträge des Datensatz 1925, anhand der Einträge in der Spalte 'Supplement'.
# Nun wird die Länge des Datensatzes, und somit die Anzahl individueller Einträge, ausgezählt.
# Durch das Abziehen der Einträge, die einen Eintrag besitzes, mit den Einträgen, die individuell sind, lässt sich feststellen, wieviele Einträge überhaupt mehrfach genannt werden.

value_counts_1925 = jobs_1925["Supplement"].value_counts()
total_count = len(value_counts_1925)
multiple_mentions = (notempty_Entries_1925 - total_count)

print ("Von diesen Einträgen sind", total_count, "individuell - es gibt also", multiple_mentions, "Einträge, die mehrfach genannt werden.")
```

```python
# Dieser Code macht ein Balkendiagramm aus den oben gewonnen Erkenntnissen.

plt.bar(["kein Eintrag", "Eintrag"], [percent_null, percent_non_null], color=['red', 'blue'])

plt.xlabel("Eintrag")
plt.ylabel("Prozent")
plt.title("Prozente zu kein Eintrag und Eintrag für Spalte 'Supplement' im Jahr 1925")

plt.show()
```

```python
# Mit diesem Code werden die Einträge, darunter auch die Berufe, ausgezählt, nach ihrer häufigsten Anzahl geordnet und die häufigsten 50 Einträge angezeigt.

print("Top 50")

print(jobs_1925["Supplement"].value_counts().head(50))
```

```python
# Es wird ein leeres Wörterbuch erstellt.
grouped_entries = {}

# Dieser Code geht nun die Spalte Supplement durch und schaut sich die ersten vier Zeichen des Strings an.
# Gibt es bereits eine Gruppe mit denselben ersten vier Zeichen, wird der Eintrag dieser Gruppe hinzugefügt.
# Wenn nicht, wird im leeren Wörterbuch eine neue Gruppe erstellt.

for index, row in jobs_1925.iterrows():
    supplement = row['Supplement']
    if pd.notna(supplement):
        key = supplement[:4]
        if key in grouped_entries:
            grouped_entries[key].append((index, supplement))
        else:
            grouped_entries[key] = [(index, supplement)]

# Dieser Code zählt nun die einzelnen Einträge in jeder Gruppe aus und zeigt sie auf.
for key, values in grouped_entries.items():
    print(f"Gruppe: {key}")
    total_count = 0
    for index, supplement in values:
        #print(f"Index: {index}, Supplement: {supplement}") Dieser Code würde jeden Eintrag in der Gruppe anzeigen, dies ist jedoch mehr verwirrend als hiflreich.
        total_count += 1
    print(f"Einträge insgesamt: {total_count}")
    print()
```
